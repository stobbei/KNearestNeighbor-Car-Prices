### Creating the predictor Function:

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
def knn_train_test(train_col, targ_col, df):
    
    perm_id = np.random.permutation(df.index)
    perm_df = df.copy().loc[perm_id]
    
    rows = len(df)
    range_train = round(0.8 * rows)
    range_test = len(df) - range_train
    train_df = perm_df.iloc[0:range_train]
    test_df = perm_df.iloc[-range_test:]
    
    knn = KNeighborsRegressor()
    knn.fit(train_df[[train_col]], train_df[targ_col])
    prediction = knn.predict(test_df[[train_col]])
    
    mse = mean_squared_error(test_df[targ_col], prediction)
    rmse = mse**0.5
    
    return rmse
```

```python
knn_train_test('horsepower', 'price', numeric_cars)
```

```python
rmse_vals = []
for col in numeric_cars.drop('price', axis=1).columns:
    rmse = knn_train_test(col, 'price', numeric_cars)
    rmse_vals.append((col, rmse))
```

```python
rmse_vals
```

Using the default k value of 5, we can see from the list above that 'horsepower' did best at predicting price.  With a root-mean-squared-error of 0.07.

How about with some other values of k?

```python
def knn_train_test(train_col, targ_col, k, df):
    
    perm_id = np.random.permutation(df.index)
    perm_df = df.copy().loc[perm_id]
    
    rows = len(df)
    range_train = round(0.8 * rows)
    range_test = len(df) - range_train
    train_df = perm_df.iloc[0:range_train]
    test_df = perm_df.iloc[-range_test:]
    
    knn = KNeighborsRegressor(n_neighbors = k)
    knn.fit(train_df[[train_col]], train_df[targ_col])
    prediction = knn.predict(test_df[[train_col]])
    
    mse = mean_squared_error(test_df[targ_col], prediction)
    rmse = mse**0.5
    
    return rmse
```

```python
rmse_vals_k = []
for col in numeric_cars.drop('price', axis=1).columns:
    for i in range(1,11,2):
        rmse = knn_train_test(col, 'price', i, numeric_cars)
        rmse_vals_k.append((col, i, rmse))
    
```

Adjusting the knn_train_test function to accept lists of columns for multivariate knn regression.  Using only the default KNN k value of 5, I'll construct multivariate regressors using the best: 2, 3, 4, and 5 features.

Ranked in order from best to worst (from the chart above):
highway_mpg, city_mpg, compression_ratio, width and horsepower

```python
def knn_train_test(train_cols, df, targ_col = 'price', k=5):
    np.random.seed(100)
    perm_id = np.random.permutation(df.index)
    perm_df = df.copy().loc[perm_id]
    
    rows = len(df)
    range_train = round(0.8 * rows)
    range_test = len(df) - range_train
    train_df = perm_df.iloc[0:range_train]
    test_df = perm_df.iloc[-range_test:]
    
    knn = KNeighborsRegressor(n_neighbors = k)
    knn.fit(train_df[train_cols], train_df[targ_col])
    prediction = knn.predict(test_df[train_cols])
    
    mse = mean_squared_error(test_df[targ_col], prediction)
    rmse = mse**0.5
    
    return rmse
```

```python
categories = ['highway_mpg', 'city_mpg', 'compression_ratio', 'width', 'horsepower']
cat_rmse_vals = []

for cat in [categories[:i] for i in range(2, 6)]:
    rmse = knn_train_test(cat, numeric_cars)
    cat_rmse_vals.append((cat, rmse))
```

```python
cat_rmse_vals
```

Using the top three models here: 2 categories, 4 categories and 5 categories, and varrying k from 1 to 25

```python
[categories[:i] for i in [2, 4, 5]]
```

```python
rmse_multi_vals = []
for cat in [categories[:i] for i in [2, 4, 5]]:
    for j in range(1,26):
        rmse = knn_train_test(cat, numeric_cars, k=j)
        rmse_multi_vals.append((cat, j, rmse))
```

```python
df = pd.DataFrame(rmse_multi_vals)
```

Lowest overall rmse occured for the 5 category regressor with k=3, with an RMSE of 0.061.  For the 4 category regressor the best model occurs for k=2 with an RMSE of 0.089.  Finally the two category KNN peaked at k=2 with an RMSE of 0.091.

Now I'll move on to implementing a k-fold cross validation model. In which the dataframe is split into a number of folds, and the RMSE value of each fold used as the test_df is averaged.

### K-Fold Cross Validation with KNeighbors Regressor:

```python
from sklearn.model_selection import cross_val_score, KFold
```

```python
kf = KFold(10, shuffle=True, random_state=100)
knn = KNeighborsRegressor()
mses = cross_val_score(knn, numeric_cars[categories], numeric_cars['price'], scoring="neg_mean_squared_error",
                      cv=kf)
```

```python
rmses = np.abs(mses)**0.5
np.mean(rmses)
```
